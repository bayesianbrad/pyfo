```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x, beta, up_shift=0, scale=1):
    """
    If we set beta suitably large then we can approximate a verticla line at the discountinuity 
    representing something continous. 
    :param x: 
    :param beta: 
    :param up_shift: 
    :param scale: 
    :return: 
    """
    return scale* (1 / (1 + np.exp(-beta*x))) + up_shift
```
